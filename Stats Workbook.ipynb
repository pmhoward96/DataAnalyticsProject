```python
#Read in passing stats CSV file
passingdf = pd.read_csv('PassingStats.csv')
passingdf.head()
```

```python
#Change object to integer
passingdf['Passing Yards'] = pd.to_numeric(passingdf['Passing Yards'])
```

```python
rushingdf = rushingdf.drop(['Passing Plays over 20 Yards', 'Passing Plays over 40 Yards'], axis = 1)
```

```python
rushingdf = rushingdf.replace('--', 0)
rushingdf['Rushing Yards'] = rushingdf['Rushing Yards'].str.replace(",", '')
rushingdf['Longest Rushing Play'] = rushingdf['Longest Rushing Play'].str.replace("T", '')
rushingdf['Rushing Yards'] = pd.to_numeric(rushingdf['Rushing Yards'])
rushingdf['Average Rushing Yards Gained'] = pd.to_numeric(rushingdf['Average Rushing Yards Gained'])
rushingdf['Longest Rushing Play'] = pd.to_numeric(rushingdf['Longest Rushing Play'])
rushingdf['Rushing First Downs'] = pd.to_numeric(rushingdf['Rushing First Downs'])
rushingdf['Rushing First Down Percentage'] = pd.to_numeric(rushingdf['Rushing First Down Percentage'])
rushingdf['Fumbles'] = pd.to_numeric(rushingdf['Fumbles'])
```

```python
rushingdf.info()
rushingdf.shape
```

```python
df = pd.merge(passingdf, rushingdf, how = 'left', left_on = ['Year', 'Player Name'], right_on = ['Year', 'Player Name'])
df.info()
```

```python
df.drop(['Unnamed: 0_x', 'Unnamed: 0_y', 'Team Name_y', 'Position_y'], axis = 1)
```

```python
df.rename(columns={'Team Name_x': 'Team Name'}, inplace=True)
df.rename(columns={'Position_x': 'Position'}, inplace=True)
```

```python
points = []
for x in range(len(df)):
    yardPoints = df.iloc[x]['Passing Yards'] % 25
    #print(yardPoints)
    tdPoints = df.iloc[x]['Passing Touchdowns'] * 4
    #print(tdPoints)
    intPoints = df.iloc[x]['Interceptions'] * -2
    #print(intPoints)
    totalPoints = yardPoints + tdPoints + intPoints
    #print(totalPoints)
    points.append(totalPoints)
df['Passing Fantasy Points'] = points
```

```python
points = []
for x in range(len(df)):
    yardPoints = df.iloc[x]['Rushing Yards'] % 10
    tdPoints = df.iloc[x]['Rushing Touchdowns'] * 6
    fumblePoints = df.iloc[x]['Fumbles'] * -2
    totalPoints = yardPoints + tdPoints + fumblePoints
    points.append(totalPoints)
df['Rushing Fantasy Points'] = points
```

```python
def assignment(df, centroids):
    for i in centroids.keys():
        # sqrt((x1 - x2)^2 - (y1 - y2)^2)
        df['distance_from_{}'.format(i)] = (
            np.sqrt(
                (df['x'] - centroids[i][0]) ** 2
                + (df['y'] - centroids[i][1]) ** 2
            )
        )
    centroid_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
    df['closest'] = df.loc[:, centroid_distance_cols].idxmin(axis=1)
    df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_')))
    df['color'] = df['closest'].map(lambda x: colmap[x])
    return df
```
