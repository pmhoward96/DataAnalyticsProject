```python
import pandas as pd
```

```python
#Read in passing stats CSV file
passingdf = pd.read_csv('PassingStats.csv')
passingdf.head()
```

```python
passingdf.describe()
```

```python
#Identify which data types will need to be changed to perform analysis
passingdf.info()
```

```python
#Replace "--" values so you can use pandas to wrangle the data
passingdf = passingdf.replace("--", '0')
```

```python
#String Replace the commas and the T's so that we can succesfully use pandas to switch to numeric value
passingdf['Passing Yards'] = passingdf['Passing Yards'].str.replace(",", '')
passingdf['Longest Passing Play'] = passingdf['Longest Passing Play'].str.replace("T", '')
```

```python
passingdf['Passing Yards Per Game'] = pd.to_numeric(passingdf['Passing Yards Per Game'])
passingdf['Passing First Down Percentage'] = pd.to_numeric(passingdf['Passing First Down Percentage'])
passingdf['Passing Completion Percentage'] = pd.to_numeric(passingdf['Passing Completion Percentage'])
passingdf['Longest Passing Play'] = pd.to_numeric(passingdf['Longest Passing Play'])
passingdf['Average Passing Yards Gained'] = pd.to_numeric(passingdf['Average Passing Yards Gained'])
```

```python
#Change object to integer
passingdf['Passing Yards'] = pd.to_numeric(passingdf['Passing Yards'])
```

```python
passingdf.info()
```

```python
#Show only stats from quarterbacks (drop all positions that are not quarterbacks)
passingdf = passingdf.drop(passingdf[passingdf.Position != 'QB'].index)
passingdf
```

```python
#Shows the top 50 quarterbacks in history with regards to average number of touchdowns per season
passingdf.groupby('Player Name')['Passing Touchdowns'].mean().sort_values(ascending= False).head(50)
```

```python
#Shows the top 50 quarterbacks in history with regards to number of touchdowns in their career to date
passingdf.groupby('Player Name')['Passing Touchdowns'].sum().sort_values(ascending= False).head(50)
```

```python
#Shows the top 50 quarterbacks in history with regards to average number of passing yards per season
passingdf.groupby('Player Name')['Passing Yards'].mean().sort_values(ascending= False).head(50)
```

```python
#Shows the top 50 quarterbacks in history with regards to number of passing yards in their career to date
passingdf.groupby('Player Name')['Passing Yards'].sum().sort_values(ascending= False).head(50)
```

```python
#Shows the top 50 quarterbacks in history with regards to average number of passing yards per game in their career to date
passingdf.groupby('Player Name')['Passing Yards Per Game'].mean().sort_values(ascending= False).head(50)
```

```python
#Creates a seperate dataframe for veteran QBs (QBs who have thrown more than 250 passes)
veteranpass = passingdf[passingdf['Passing Attempts'] > 250]
```

```python
#Shows the top 50 veteran QBs with regards to their average completition percentage per season
veteranpass.groupby('Player Name')['Passing Completion Percentage'].mean().sort_values(ascending= False).head(50)
```

```python
#Shows the top 50 veteran QBs with regards to the total number of passes they have attempted in their career
veteranpass.groupby('Player Name')['Passing Attempts'].sum().sort_values(ascending= False).head(50)
```

```python
#Shows the top 50 veteran QBs with regards to their average completition percentage per season
veteranpass.groupby('Player Name')['Interceptions'].sum().sort_values(ascending= False).head(50)
```
