```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read in passing stats CSV file
passingdf = pd.read_csv('PassingStats.csv')
passingdf.head()
```

```python
passingdf.describe()
```

```python
#Identify which data types will need to be changed to perform analysis
passingdf.info()
```

```python
#Replace "--" values so you can use pandas to wrangle the data
passingdf = passingdf.replace("--", '0')
```

```python
#String Replace the commas and the T's so that we can succesfully use pandas to switch to numeric value
passingdf['Passing Yards'] = passingdf['Passing Yards'].str.replace(",", '')
passingdf['Longest Passing Play'] = passingdf['Longest Passing Play'].str.replace("T", '')
```

```python
passingdf['Passing Yards Per Game'] = pd.to_numeric(passingdf['Passing Yards Per Game'])
passingdf['Passing First Down Percentage'] = pd.to_numeric(passingdf['Passing First Down Percentage'])
passingdf['Passing Completion Percentage'] = pd.to_numeric(passingdf['Passing Completion Percentage'])
passingdf['Longest Passing Play'] = pd.to_numeric(passingdf['Longest Passing Play'])
passingdf['Average Passing Yards Gained'] = pd.to_numeric(passingdf['Average Passing Yards Gained'])
```

```python
#Change object to integer
passingdf['Passing Yards'] = pd.to_numeric(passingdf['Passing Yards'])
```

```python
passingdf.info()
```

```python
#Show only stats from quarterbacks (drop all positions that are not quarterbacks)
passingdf = passingdf.drop(passingdf[passingdf.Position != 'QB'].index)
passingdf
```

```python
passingdf.groupby('Player Name')['Quaterback Rating'].mean().sort_values(ascending= False).head(5)
```

```python
#Shows the top 50 quarterbacks in history with regards to average number of touchdowns per season
passingdf.groupby('Player Name')['Passing Touchdowns'].mean().sort_values(ascending= False).head(5)
```

```python
#Shows the top 50 quarterbacks in history with regards to number of touchdowns in their career to date
passingdf.groupby('Player Name')['Passing Touchdowns'].sum().sort_values(ascending= False).head(5)
```

```python
#Shows the top 50 quarterbacks in history with regards to average number of passing yards per season
passingdf.groupby('Player Name')['Passing Yards'].mean().sort_values(ascending= False).head(5)
```

```python
#Shows the top 50 quarterbacks in history with regards to number of passing yards in their career to date
passingdf.groupby('Player Name')['Passing Yards'].sum().sort_values(ascending= False).head(5)
```

```python
#Shows the top 50 quarterbacks in history with regards to average number of passing yards per game in their career to date
passingdf.groupby('Player Name')['Passing Yards Per Game'].mean().sort_values(ascending= False).head(5)
```

```python
points = []
for x in range(len(passingdf)):
    yardPoints = passingdf.iloc[x]['Passing Yards'] % 25
    #print(yardPoints)
    tdPoints = passingdf.iloc[x]['Passing Touchdowns'] * 4
    #print(tdPoints)
    intPoints = passingdf.iloc[x]['Interceptions'] * -2
    #print(intPoints)
    totalPoints = yardPoints + tdPoints + intPoints
    #print(totalPoints)
    points.append(totalPoints)
passingdf['Passing Fantasy Points'] = points
passingdf
```

```python
from sklearn import linear_model
import statsmodels.api as sm
X = passingdf[['Passing Touchdowns',
               'Passing Attempts',
               'Passing Completions',
               'Passing Completion Percentage',
               'Passing First Down Percentage',
               'Interceptions', 
               'Passing Yards Per Game', 
               'Passing First Downs',
               'Sacks',
               'Quaterback Rating',
               'Longest Passing Play',
               'Passing Plays over 20 Yards',
               'Passing Plays over 40 Yards',
               'Year'
               
              ]] # here we have 2 variables for multiple regression. If you just want to use one variable for simple linear regression, then use X = df['Interest_Rate'] for example.Alternatively, you may add additional variables within the brackets
Y = passingdf['Passing Fantasy Points']
 
# with sklearn
regr = linear_model.LinearRegression()
regr.fit(X, Y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)




# with statsmodels
X = sm.add_constant(X) # adding a constant
 
model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 
 
print_model = model.summary()
print(print_model)
```
